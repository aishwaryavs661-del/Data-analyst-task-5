```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# Show the first 5 rows
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set a clean grid style for mobile viewing
sns.set_theme(style="whitegrid")

# Create a histogram for the 'Age' column
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='Age', kde=True, color='skyblue')
plt.title('Distribution of Passenger Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Fare', color='salmon')
plt.title('Boxplot of Passenger Fare (Outlier Detection)')
plt.xlabel('Ticket Fare')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
# Create a bar plot comparing Class, Survival Rate, and Sex
sns.barplot(data=df, x='Pclass', y='Survived', hue='Sex', palette='muted', ci=None)
plt.title('Survival Rate by Passenger Class and Gender')
plt.ylabel('Survival Probability')
plt.xlabel('Passenger Class (1 = 1st, 2 = 2nd, 3 = 3rd)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Select only the columns that contain numeric data
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Generate the heatmap matrix
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Titanic Variables')
plt.show()
```
